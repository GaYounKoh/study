# 결측치 확인
```
결측치 없으므로 처리과정 pass
```

````python
def check_missing_col(df):
    missing_col = []
    for col in df.columns:
        missing_values = sum(df[col].isna())
        is_missing = True if missing_values >= 1 else False
        if is_missing:
            print(f'결측치 있는 col: {col}')
            print(f'해당 col에 총 {missing_values}개의 결측치 존재')
            missing_col.append([col, df[col].dtype])
    if missing_col == []:
        print('결측치 없음.')
    return missing_col

missing_col = check_missing_col(train)
````

````python
# 전체의 데이터 타입 확인
# http://daplus.net/python-python-pandas%EC%97%90%EC%84%9C-%EC%97%B4%EC%9D%98-dtype%EC%9D%84-%ED%99%95%EC%9D%B8%ED%95%98%EB%8A%94-%EB%B0%A9%EB%B2%95/
# https://pandas.pydata.org/docs/reference/api/pandas.api.types.is_numeric_dtype.html#
# 수치형 해결이 도저히 안돼서 읽어봄.. (두 번째 링크 -> source)
# pandas는 재귀함수로 함수 만듦.

#결국 numpy.dtype attribute 구글링 해서 해결함.
# data.dtype.name하면 그 객체 이름 문자형으로 반환해줌.
# 기억해야할 부분
# data.dtype.values
# data.dtype.index
# data.dtype.name
#https://numpy.org/doc/stable/reference/arrays.dtypes.html


num_feat = []
cat_feat = []
# train.dtypes#.values
# train.dtypes=='O'
for i,j in enumerate(train.dtypes.values):
    k = j.name#, j.str
    if k.startswith("i") or k.startswith("f"):
        # print(i,k)
        num_feat.append(train.dtypes.index[i])
    else:
        # print(i,k)
        cat_feat.append(train.dtypes.index[i])
    # if j == 'int64':
    #     print(i)
        
# np.where(train.dtypes.values == 'int64')

````

# 상관관계 시각화
```
feature들 간의 상관관계
전체 feature들 간의 상관관계 살펴보고
주요 feat과 target feat의 상관관계를 산점도를 통해 살펴보겠음.
```

````python
# 먼저 상관계수 계산을 위해
# 텍스트 형식의 데이터를 숫자로 변환.
from sklearn.preprocessing import LabelEncoder

corr_df = train.iloc[:,1:].copy()
corr_df[corr_df.columns[corr_df.dtypes == 'O']] = corr_df[corr_df.columns[corr_df.dtypes == 'O']].astype(str).apply(LabelEncoder().fit_transform)
# dtype이 object형인 열들이 바뀌게 됨.
````

````python
# train.iloc[:,1:]
````

````python
# 상관관계 분석도
_= plt.figure(figsize = (15,10))

heat_table = corr_df.corr()
mask = np.zeros_like(heat_table)
mask[np.triu_indices_from(mask)] = True # 위쪽 덮음.
heatmap_ax = sns.heatmap(heat_table, annot = True, mask = mask, cmap = 'coolwarm')
_= heatmap_ax.set_xticklabels(heatmap_ax.get_xticklabels(), fontsize = 15, rotation = 45)
_= heatmap_ax.set_yticklabels(heatmap_ax.get_yticklabels(), fontsize = 15)
_= plt.title('correlation between features', fontsize = 40)
plt.show()
````

## target과 수치형 feat의 관계

````python
num_feat[1:]
````

````python
feat = num_feat[1:]

_= plt.figure(figsize = (20,15))
_= plt.suptitle("Scatter Plot", fontsize = 40)

# target은 제외하고 시각화
for i in range(len(feat[:-1])):
    _= plt.subplot(4,3,i+1)
    _= plt.xlabel(feat[i])
    _= plt.ylabel(target)
    corr_score = train.iloc[:,1:][[feat[i], target]].corr().iloc[0,1].round(2)
    c = 'red' if corr_score > 0 else 'blue'

    _= plt.scatter(train.iloc[:,1:][feat[i]], train.iloc[:,1:][target], color = c, label = f"corr = {corr_score}")
    _= plt.legend(fontsize = 15)
_= plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()






````

```
상관관계 그래프를 보니 몇몇 변수가 타겟변수와 상당한 양의 상관관계를 보이고 있음을 확인할 수 있음.
```
