## 반올림

```python
# pd.DataFrame(map(lambda x: round(x) if x != 'NaN' else 'N', train['hour_bef_temperature']))
for i in tr_cols:
    train[i] = pd.DataFrame(map(lambda x: int(round(x)) if x >= 1 else round(x, 2), train[i])) # 1 미만일 때는 x그대로
```

```python
train.info()
```

```python
## count 별~~
pd.pivot_table(train, index = ['count'])
```
